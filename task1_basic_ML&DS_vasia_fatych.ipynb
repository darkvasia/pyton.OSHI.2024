```python
def solve_quad(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return None
    elif a == 0:
        return None
    elif discriminant == 0:
        root = -b / (2*a)
        return root,
    else:
        root1 = (-b + (discriminant)**0.5) / (2*a)
        root2 = (-b - (discriminant)**0.5) / (2*a)
        return root1, root2
roots = solve_quad(a=1, b=-2, c=1)
if roots is not None:
    print(*roots)
roots = solve_quad(a=1, b=0, c=-1)
if roots is not None:
    print(*roots)
roots = solve_quad(a=0, b=2, c=-1)
print(roots)
```

```python
def merge(a, b):
    result = [val for pair in zip(a, b) for val in pair] + a[len(b):] + b[len(a):]
    return sorted(result)
arr1 = [1, 3, 3, 5]
arr2 = [3, 5, 5, 7]
result_arr = merge(arr1, arr2)
print(result_arr)  
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    left_half, right_half = map(merge_sort, [arr[:len(arr)//2], arr[len(arr)//2:]])
    result = []
    i = j = 0
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            result.append(left_half[i])
            i += 1
        else:
            result.append(right_half[j])
            j += 1
    result += left_half[i:]
    result += right_half[j:]
    return result
arr = [3, 1, 3, 5]
sorted_arr = merge_sort(arr)
print(*sorted_arr)

```

```python
import numpy as np
def are_arrays_equal(X, Y, Z):
    arrays = [X, Y, Z]
    return np.all(np.array_equal(arrays, arrays[1:] + [arrays[0]]))
X = np.array([1, 2, 3])
Y = np.array([1, 2, 3])
Z = np.array([1, 2, 3])
result = are_arrays_equal(X, Y, Z)
print(result)
```

```python
import pandas as pd
def count_sex(filepath):
    data = pd.read_csv(filepath)
    male_count = (data['Sex'] == 'male').sum()
    female_count = (data['Sex'] == 'female').sum()
    return male_count, female_count
def compare_sex(filepath):
    male_count, female_count = count_sex(filepath)
    if male_count > female_count:
        return "more man"
    elif male_count < female_count:
        return "more woman"
    else:
        return "men and women equally"
result = compare_sex("./titanic.csv")
print(result)
```

```python
import pandas as p
def passenger_stats(filepath):
    data = pd.read_csv(filepath)
    avg_age = data['Age'].mean()
    males = data[data['Sex'] == 'male']
    females = data[data['Sex'] == 'female']
    avg_age_males = males['Age'].mean()
    avg_age_females = females['Age'].mean()
    max_age_males, min_age_males = males['Age'].max(), males['Age'].min()
    max_age_females, min_age_females = females['Age'].max(), females['Age'].min()
    return avg_age, avg_age_males, avg_age_females, max_age_males, min_age_males, max_age_females, min_age_females
avg_age, avg_age_males, avg_age_females, max_age_males, min_age_males, max_age_females, min_age_females = passenger_stats('./titanic.csv')
print(f"Середній вік пасажирів: {avg_age:.2f}")
print(f"Середній вік чоловіків: {avg_age_males:.2f}")
print(f"Середній вік жінок: {avg_age_females:.2f}")
print(f"Максимальний вік серед чоловіків: {max_age_males}")
print(f"Мінімальний вік серед чоловіків: {min_age_males}")
print(f"Максимальний вік серед жінок: {max_age_females}")
print(f"Мінімальний вік серед жінок: {min_age_females}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
functions_dict = {
    r'$\frac{\sin(x)}{x}$': lambda x: np.sin(x) / x,
    r'$\frac{\cos(x)}{x^2 + 1}$': lambda x: np.cos(x) / (x**2 + 1),
    r'$\cos^2(x) - \frac{\sin(x)}{2}$': lambda x: np.cos(x)**2 - np.sin(x)/2,
    r'$\cos(x)$': np.cos,
    r'$\frac{\sin(x)}{x^2 + 1}$': lambda x: np.sin(x) / (x**2 + 1)
}
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(10, 6))
for label, func in functions_dict.items():
    plt.plot(x, func(x), label=label)
plt.title('Графіки функцій')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pl
def generate_random_data(size=1000, loc=2.0, scale=3.5):
    return np.random.normal(loc=loc, scale=scale, size=size)
def plot_histogram(data, bins=30, color='blue', edgecolor='black', alpha=0.7, density=True, label=None):
    plt.hist(data, bins=bins, color=color, edgecolor=edgecolor, alpha=alpha, density=density, label=label)


random_data = generate_random_data()
plt.figure(figsize=(10, 6))
plot_histogram(random_data, label='Гістограма щільності розподілу')
plt.title('Щільність розподілу випадкової змінної')
plt.xlabel('Значення')
plt.ylabel('Щільність розподілу')
plt.legend()
plt.show()
```
