```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

task1

```python
n_points_per_class = 500
mean_class_1 = [2, 2]
mean_class_2 = [-2, -2]

cov_matrix_class_1 = [[0.8, 0.1], [0.1, 0.8]]
cov_matrix_class_2 = [[0.8, 0.1], [0.1, 0.8]]

class_1_points = np.random.multivariate_normal(mean_class_1, cov_matrix_class_1, n_points_per_class)
class_1_labels = np.ones(n_points_per_class)

class_2_points = np.random.multivariate_normal(mean_class_2, cov_matrix_class_2, n_points_per_class)
class_2_labels = np.zeros(n_points_per_class)

data_points = np.vstack((class_1_points, class_2_points))
labels = np.hstack((class_1_labels, class_2_labels))

data = pd.DataFrame(data_points, columns=["x", "y"])
data["class"] = labels

csv_path = "linear_separable_dataset.csv"
data.to_csv(csv_path, index=False)

plt.figure(figsize=(10, 6))
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], color='blue', label='Class 1', alpha=0.6)
plt.scatter(class_2_points[:, 0], class_2_points[:, 1], color='red', label='Class 2', alpha=0.6)

x_vals = np.linspace(-5, 5, 100)
y_vals = x_vals
plt.plot(x_vals, y_vals, 'k--', label='Separation Line')

plt.title('Linear Separable Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(class_1_points[:, 0], bins=20, color='blue', alpha=0.7, label='Class 1 x')
plt.hist(class_1_points[:, 1], bins=20, color='green', alpha=0.7, label='Class 1 y')
plt.title('Class 1 Distribution')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(class_2_points[:, 0], bins=20, color='red', alpha=0.7, label='Class 2 x')
plt.hist(class_2_points[:, 1], bins=20, color='orange', alpha=0.7, label='Class 2 y')
plt.title('Class 2 Distribution')
plt.legend()

plt.tight_layout()
plt.show()


```

```python
n_points_per_class = 300
mean_class_1 = [2, 2]
mean_class_2 = [-2, -2]
mean_class_3 = [2, -2]

cov_matrix_class_1 = [[1.5, 0.5], [0.5, 1.5]]
cov_matrix_class_2 = [[1.5, 0.5], [0.5, 1.5]]
cov_matrix_class_3 = [[1.5, 0.5], [0.5, 1.5]]

class_1_points = np.random.multivariate_normal(mean_class_1, cov_matrix_class_1, n_points_per_class)
class_1_labels = np.ones(n_points_per_class)

class_2_points = np.random.multivariate_normal(mean_class_2, cov_matrix_class_2, n_points_per_class)
class_2_labels = np.full(n_points_per_class, 2)

class_3_points = np.random.multivariate_normal(mean_class_3, cov_matrix_class_3, n_points_per_class)
class_3_labels = np.full(n_points_per_class, 3)

data_points = np.vstack((class_1_points, class_2_points, class_3_points))
labels = np.hstack((class_1_labels, class_2_labels, class_3_labels))

data = pd.DataFrame(data_points, columns=["x", "y"])
data["class"] = labels

csv_path = "data.csv"
data.to_csv(csv_path, index=False)

plt.figure(figsize=(10, 6))
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], color='blue', label='Class 1', alpha=0.6)
plt.scatter(class_2_points[:, 0], class_2_points[:, 1], color='red', label='Class 2', alpha=0.6)
plt.scatter(class_3_points[:, 0], class_3_points[:, 1], color='green', label='Class 3', alpha=0.6)

plt.title('Overlapping Classes Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.hist(class_1_points[:, 0], bins=20, color='blue', alpha=0.7, label='Class 1 x')
plt.hist(class_1_points[:, 1], bins=20, color='green', alpha=0.7, label='Class 1 y')
plt.title('Class 1 Distribution')
plt.legend()

plt.subplot(3, 2, 2)
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], color='blue', alpha=0.6)
plt.title('Class 1 Correlation')

plt.subplot(3, 2, 3)
plt.hist(class_2_points[:, 0], bins=20, color='red', alpha=0.7, label='Class 2 x')
plt.hist(class_2_points[:, 1], bins=20, color='orange', alpha=0.7, label='Class 2 y')
plt.title('Class 2 Distribution')
plt.legend()

plt.subplot(3, 2, 4)
plt.scatter(class_2_points[:, 0], class_2_points[:, 1], color='red', alpha=0.6)
plt.title('Class 2 Correlation')

plt.subplot(3, 2, 5)
plt.hist(class_3_points[:, 0], bins=20, color='green', alpha=0.7, label='Class 3 x')
plt.hist(class_3_points[:, 1], bins=20, color='yellow', alpha=0.7, label='Class 3 y')
plt.title('Class 3 Distribution')
plt.legend()

plt.subplot(3, 2, 6)
plt.scatter(class_3_points[:, 0], class_3_points[:, 1], color='green', alpha=0.6)
plt.title('Class 3 Correlation')
plt.tight_layout()
plt.show()

```

task 3

```python
n_points_per_class = 500
inner_radius = 2
inner_width = 0.5
outer_radius = 5
outer_width = 0.8

theta_inner = np.random.uniform(0, 2 * np.pi, n_points_per_class)
r_inner = np.random.normal(inner_radius, inner_width, n_points_per_class)
x_inner = r_inner * np.cos(theta_inner)
y_inner = r_inner * np.sin(theta_inner)
class_1_points = np.column_stack((x_inner, y_inner))
class_1_labels = np.ones(n_points_per_class)

theta_outer = np.random.uniform(0, 2 * np.pi, n_points_per_class)
r_outer = np.random.normal(outer_radius, outer_width, n_points_per_class)
x_outer = r_outer * np.cos(theta_outer)
y_outer = r_outer * np.sin(theta_outer)
class_2_points = np.column_stack((x_outer, y_outer))
class_2_labels = np.zeros(n_points_per_class)

data_points = np.vstack((class_1_points, class_2_points))
labels = np.hstack((class_1_labels, class_2_labels))

data = pd.DataFrame(data_points, columns=["x", "y"])
data["class"] = labels

csv_path = "data.csv"
data.to_csv(csv_path, index=False)

plt.figure(figsize=(8, 8))
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], color='blue', label='Inner Torus (Class 1)', alpha=0.6)
plt.scatter(class_2_points[:, 0], class_2_points[:, 1], color='red', label='Outer Torus (Class 2)', alpha=0.6)
plt.title('Torus in Torus Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(r_inner, bins=20, color='blue', alpha=0.7, label='Inner Torus Radius')
plt.title('Inner Torus Radius Distribution')
plt.legend()
plt.subplot(1, 2, 2)
plt.hist(r_outer, bins=20, color='red', alpha=0.7, label='Outer Torus Radius')
plt.title('Outer Torus Radius Distribution')
plt.legend()
plt.tight_layout()
plt.show()


```

```python
n_points_per_class = 500
noise = 0.2

theta_class_1 = np.sqrt(np.random.rand(n_points_per_class)) * 2 * np.pi * 3  
r_class_1 = theta_class_1 + np.random.normal(0, noise, n_points_per_class)
x_class_1 = r_class_1 * np.cos(theta_class_1)
y_class_1 = r_class_1 * np.sin(theta_class_1)
class_1_points = np.column_stack((x_class_1, y_class_1))
class_1_labels = np.ones(n_points_per_class)

theta_class_2 = np.sqrt(np.random.rand(n_points_per_class)) * 2 * np.pi * 3
r_class_2 = -theta_class_2 + np.random.normal(0, noise, n_points_per_class)
x_class_2 = r_class_2 * np.cos(theta_class_2)
y_class_2 = r_class_2 * np.sin(theta_class_2)
class_2_points = np.column_stack((x_class_2, y_class_2))
class_2_labels = np.zeros(n_points_per_class)


data_points = np.vstack((class_1_points, class_2_points))
labels = np.hstack((class_1_labels, class_2_labels))

data = pd.DataFrame(data_points, columns=["x", "y"])
data["class"] = labels

csv_path = "data.csv"
data.to_csv(csv_path, index=False)
print(f"    {csv_path}")


plt.figure(figsize=(8, 8))
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], color='blue', label='Spiral 1 (Class 1)', alpha=0.6)
plt.scatter(class_2_points[:, 0], class_2_points[:, 1], color='red', label='Spiral 2 (Class 2)', alpha=0.6)
plt.title('Double Spiral Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], color='blue', alpha=0.6)
plt.title('Spiral 1 (Class 1)')
plt.subplot(1, 2, 2)
plt.scatter(class_2_points[:, 0], class_2_points[:, 1], color='red', alpha=0.6)
plt.title('Spiral 2 (Class 2)')
plt.tight_layout()
plt.show()
```
